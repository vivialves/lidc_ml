```python
def extract_nodule_characteristics(xml_file_path):
    """
    Extracts patient, radiologist, nodule ID, and characteristics from a LIDC-IDRI XML file.

    Args:
        xml_file_path (str): The path to the XML file.

    Returns:
        pandas.DataFrame: A DataFrame containing the extracted information.
                          Returns an empty DataFrame if no nodule information is found.
    """
    data = []
    try:
        tree = ET.parse(xml_file_path)
        root = tree.getroot()

        dirname = xml_file_path
        patient_id_element = xml_file_path.split('/')[8]
        patient_id = patient_id_element if patient_id_element is not None else None

        for reading_session in root.findall('.//{http://www.nih.gov}readingSession'):
            radiologist_id_element = reading_session.findtext('{http://www.nih.gov}servicingRadiologistID')
            radiologist_id = radiologist_id_element if radiologist_id_element else None

            for nodule in reading_session.findall('.//{http://www.nih.gov}unblindedReadNodule'):
                nodule_id_element = nodule.findtext('{http://www.nih.gov}noduleID')
                nodule_id = nodule_id_element if nodule_id_element else None

                # Initialize all characteristic variables *before* the if statement
                malignancy = None
                subtlety = None
                internal_structure = None
                calcification = None
                sphericity = None
                margin = None
                lobulation = None
                spiculation = None
                texture = None
                
                characteristics = nodule.find('{http://www.nih.gov}characteristics')
                if characteristics is not None:
                    malignancy = characteristics.findtext('{http://www.nih.gov}malignancy')
                    subtlety = characteristics.findtext('{http://www.nih.gov}subtlety')
                    internal_structure = characteristics.findtext('{http://www.nih.gov}internalStructure')
                    calcification = characteristics.findtext('{http://www.nih.gov}calcification')
                    sphericity = characteristics.findtext('{http://www.nih.gov}sphericity')
                    margin = characteristics.findtext('{http://www.nih.gov}margin')
                    lobulation = characteristics.findtext('{http://www.nih.gov}lobulation')
                    spiculation = characteristics.findtext('{http://www.nih.gov}spiculation')
                    texture = characteristics.findtext('{http://www.nih.gov}texture')

                image_z_position = None # Initialize
                list_coord = [] # Initialize
                roi = nodule.find('{http://www.nih.gov}roi')
                if roi is not None:
                    if isinstance(roi, list):
                        for roi_element in roi: # Changed roi to roi_element
                            image_z_position = roi_element.findtext('{http://www.nih.gov}imageZposition') # Corrected tag name
                            edgemap = roi_element.findall('{http://www.nih.gov}edgeMap')
                            for map_element in edgemap: # Changed map to map_element
                                coord_x = map_element.findtext('{http://www.nih.gov}xCoord')
                                coord_y = map_element.findtext('{http://www.nih.gov}yCoord')
                                list_coord.append((coord_x, coord_y))
                    else:
                            image_z_position = roi.findtext('{http://www.nih.gov}imageZposition') # Corrected tag name
                            edgemap = roi.findall('{http://www.nih.gov}edgeMap')
                            for map_element in edgemap: # Changed map to map_element
                                coord_x = map_element.findtext('{http://www.nih.gov}xCoord')
                                coord_y = map_element.findtext('{http://www.nih.gov}yCoord')
                                list_coord.append((coord_x, coord_y))

                data.append({
                    'dirname': dirname,
                    'patient_id': patient_id,
                    'radiologist_id': radiologist_id,
                    'nodule_id': nodule_id,
                    'malignancy': malignancy,
                    'subtlety': subtlety,
                    'internalStructure': internal_structure,
                    'calcification': calcification,
                    'sphericity': sphericity,
                    'margin': margin,
                    'lobulation': lobulation,
                    'spiculation': spiculation,
                    'texture': texture,
                    'image_z_position': float(image_z_position),
                    'coordinates': list_coord
                })
    except FileNotFoundError:
        print(f"Error: XML file not found at {xml_file_path}")
    except ET.ParseError as e:
        print(f"Error parsing XML file {xml_file_path}: {e}")
    return pd.DataFrame(data)
```

```python
if __name__ == "__main__":
    # xml_dir = '/path/to/your/LIDC-XML-only/tcia-lidc-xml'  # Replace with the path to your XML directory
    all_nodule_data = []

    for index, row in df.iterrows():
        filenames = row['filename']
        dirname = row['dirname']
        if isinstance(filenames, list):
            xml_filename = None
            for filename in filenames:
                if filename.endswith('.xml'):
                    xml_filename = filename
                    break
            if xml_filename:
                full_path = os.path.join(dirname, filename)
                df = extract_nodule_characteristics(full_path)
                if not df.empty:
                    all_nodule_data.append(df)

    if all_nodule_data:
        final_df = pd.concat(all_nodule_data, ignore_index=True)
        print(final_df)
        # You can now save this DataFrame to a CSV file if needed
        # final_df.to_csv('lidc_nodule_characteristics.csv', index=False)
    else:
        print("No nodule characteristics data extracted.")
```

```python
def calculate_centroid(coordinates):
    if not coordinates:
        return None, None
    x_coords = []
    y_coords = []
    for coord in coordinates:
        try:
            x_coords.append(int(coord[0]))
            y_coords.append(int(coord[1]))
        except (ValueError, TypeError):
            # Handle cases where coordinate conversion fails
            return None, None
    if x_coords and y_coords:
        return np.mean(x_coords), np.mean(y_coords)
    else:
        return None, None

def match_nodules_by_centroid(df, z_tolerance=5.0, xy_tolerance=15.0):
    df['matched_nodule_group'] = None
    group_counter = 0

    for patient_id in df['patient_id'].unique():
        patient_df = df[df['patient_id'] == patient_id].copy()
        radiologists = patient_df['radiologist_id'].unique()

        nodule_centroids = []
        nodule_indices = []
        for index, row in patient_df.iterrows():
            z = row['image_z_position']
            coords_list = row['coordinates']
            centroid_x, centroid_y = calculate_centroid(coords_list)
            if centroid_x is not None and centroid_y is not None and z is not None:
                nodule_centroids.append([centroid_x, centroid_y, z])
                nodule_indices.append(index)

        if not nodule_centroids:
            continue

        neighbors = NearestNeighbors(n_neighbors=len(nodule_centroids), algorithm='ball_tree')
        neighbors.fit(nodule_centroids)
        distances, indices = neighbors.kneighbors(nodule_centroids)

        matched = [False] * len(nodule_centroids)

        for i in range(len(nodule_centroids)):
            if not matched[i]:
                group_counter += 1
                matched[i] = True
                df.loc[nodule_indices[i], 'matched_nodule_group'] = group_counter
                cx1, cy1, z1 = nodule_centroids[i]

                for j in range(i + 1, len(nodule_centroids)):
                    if not matched[j]:
                        cx2, cy2, z2 = nodule_centroids[j]
                        xy_dist = np.sqrt((cx1 - cx2)**2 + (cy1 - cy2)**2)
                        z_diff = abs(z1 - z2)

                        if xy_dist <= xy_tolerance and z_diff <= z_tolerance:
                            matched[j] = True
                            df.loc[nodule_indices[j], 'matched_nodule_group'] = group_counter

    return df
```

```python
df_matched_centroid = match_nodules_by_centroid(final_df.copy(), z_tolerance=5.0, xy_tolerance=15.0)

print(df_matched_centroid[['patient_id', 'radiologist_id', 'nodule_id', 'image_z_position', 'coordinates', 'matched_nodule_group']].head())

if 'matched_nodule_group' in df_matched_centroid.columns and not df_matched_centroid['matched_nodule_group'].isnull().all():
    grouped_malignancy = df_matched_centroid.groupby('matched_nodule_group')['malignancy'].apply(list)
    print("\nGrouped Malignancy Ratings (by centroid) for Potential Matches:")
    print(grouped_malignancy)
else:
    print("\nNo potential matching nodules found based on centroid and tolerances.")
```
